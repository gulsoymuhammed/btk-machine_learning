```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
df=pd.read_csv("/content/drive/MyDrive/btk-makine öğrenimi/çalışmalar/bilkav.com_odev_tenis.csv")
```

```python
df
```

```python
outlook=df.iloc[:,:1].values
```

```python
outlook
```

```python
ohe=preprocessing.OneHotEncoder()
outlook=ohe.fit_transform(outlook).toarray()
```

```python
windy=df.iloc[:,3:4].values
windy
```

```python
le=preprocessing.LabelEncoder()
windy=le.fit_transform(df.iloc[:,3:4])
```

```python
windy
```

```python
play=df.iloc[:,4:5].values
play
```

```python
le=preprocessing.LabelEncoder()
play=le.fit_transform(df.iloc[:,4:5])
```

```python
play
```

```python
temp=df.iloc[:,1:3]
temp
```

```python
sonuc1=pd.DataFrame(data=outlook,index=range(14),columns=["overcast","rainy","sunny"])
sonuc1
```

```python
sonuc2=pd.DataFrame(data=windy,index=range(14),columns=["windy"])
sonuc2
```

```python
sonuc3=pd.DataFrame(data=play,index=range(14),columns=["play"])
sonuc3
```

```python
s1=pd.concat([sonuc1,sonuc2],axis=1)
s2=pd.concat([s1,sonuc3],axis=1)
s3=pd.concat([s2,temp],axis=1)
```

```python
s3
```

```python
#verilerin egitim ve test icin bolunmesi
from sklearn.model_selection import train_test_split

x_train, x_test,y_train,y_test = train_test_split(s3.iloc[:,:-1],s3.iloc[:,-1:],test_size=0.33, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

regression=LinearRegression()
regression.fit(x_train,y_train)
y_pred=regression.predict(x_test)
```
